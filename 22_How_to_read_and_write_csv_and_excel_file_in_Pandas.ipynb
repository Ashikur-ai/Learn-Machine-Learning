```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3, 4], 'B':[11, 22, 33, 4], 'C':['aaa', 'bbb', 'ccc', 'dd']})
```

```python
df
```

#Convert to CSV file

```python
df.to_csv('data')
```

```python
df.to_csv('No_index', index=False)
```

```python
df1 = pd.read_csv('data')
```

```python
df1
```

```python
df2 = pd.read_csv('No_index')
```

```python
df2
```

#Convert to excel file

```python
df.to_excel('excel.xlsx')
```

```python
df3 = pd.read_excel('excel.xlsx')
```
